```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
PATH_HISTORY = 'logs/training_history_2023-06-18_21:27:21.csv'
```

```python
df_his = pd.read_csv(PATH_HISTORY)
df_his
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].plot(df_his['epoch'], df_his['train_obj_loss'], label='train_obj_loss')
axes[0].plot(df_his['epoch'], df_his['val_obj_loss'], label='val_obj_loss')
axes[0].set_title('Object Detection Loss')
axes[0].set_xlabel('Epochs')
axes[0].set_ylabel('Loss')
axes[0].legend()

axes[1].plot(df_his['epoch'], df_his['train_seg_loss'], label='train_seg_loss')
axes[1].plot(df_his['epoch'], df_his['val_seg_loss'], label='val_seg_loss')
axes[1].set_title('Semantic Segmentation Loss')
axes[1].set_xlabel('Epochs')
axes[0].set_ylabel('Loss')
axes[1].legend()
```
